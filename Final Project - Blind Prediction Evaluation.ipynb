```python
import pandas as pd
import yfinance as yf
from datetime import datetime, timedelta
from dateutil.relativedelta import relativedelta

end_date = datetime.now() + relativedelta(days=1)
start_date = end_date - relativedelta(days=1)
end_date_formatted = str(end_date).split(" ")[0]
start_date_formatted = str(start_date).split(" ")[0]

filename = "FinalProject-FutureEstimation_" + start_date_formatted + ".csv"  #2022-04-27
df = pd.read_csv(filename)
df
```

```python
# Let's pull in those actual values.
from tqdm.auto import tqdm
actual_close = []
progress_bar = tqdm(
        range(df.shape[0])
    )
for symbol in df['Symbol'].values:
    print(symbol)
    data = yf.download(symbol, start=start_date_formatted, end=end_date_formatted)
    current = data.loc[start_date_formatted]
    actual_close.append(current['Close'])
    progress_bar.update(1)
df['True Close'] = actual_close
df
```

```python
yesterday_close = df['Last Known Close'].values
did_stock_go_up_actual = []
for i in range(0,len(yesterday_close)):
    if(yesterday_close[i] < actual_close[i]):
        did_stock_go_up_actual.append('↑')
    else:
        did_stock_go_up_actual.append('↓')
df['True Stock Direction'] = did_stock_go_up_actual
df
```

```python
#Okay, now we've got our eval dataframe!  Let's do some analysis!
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
import numpy as np

pred_close = df['Estimated Close'].values
actual_close = df['True Close'].values
%pylab inline
x=np.linspace(0,4000,4001) 

plt.scatter(actual_close, pred_close)
plt.plot(x,x,'k-') # identity line
plt.title(" MSE: " + str(mean_squared_error(actual_close, pred_close)))
plt.xlabel("Stock Actual")
plt.ylabel("Stock Predicted")
plt.show()

ax3 = plt.subplot()
ax3.scatter(actual_close, pred_close)
ax3.set_xlabel("Stock Predicted")
ax3.set_ylabel("Stock Actual")
ax3.set_title('Zoomed in')
ax3.set_ylim(bottom=0)
ax3.set_xlim(left=0)
ax3.set_ylim(top=800)
ax3.set_xlim(right=800)
plt.plot(x,x,'k-') # identity line
```

```python
# Woah, looks like we tended to underestimate pretty consistently.  That's probably better than the other way around, though.
# Let's take a look at those R1 scores.
did_stock_go_up_actual = df['True Stock Direction'].values
did_stock_go_up_predicted = df['Stock Direction'].values
stock_went_up_pred = list(map(lambda x: 1 if (x=='↑') else 0, did_stock_go_up_predicted))
stock_went_up_actual =list(map(lambda x: 1 if (x=='↑') else 0, did_stock_go_up_actual))
from sklearn import metrics
print(f"F1 Score: {metrics.f1_score(stock_went_up_actual, stock_went_up_pred)}.")
fpr, tpr, threshholds = metrics.roc_curve(stock_went_up_actual,stock_went_up_pred,pos_label=1)
base_fpr, base_tpr, _ = metrics.roc_curve(stock_went_up_actual, [1 for _ in range(len(stock_went_up_actual))])
plt.figure(figsize = (8, 6))
plt.rcParams['font.size'] = 16
plt.plot(fpr, tpr, 'r', label = 'model')
plt.plot(base_fpr, base_tpr, 'b', label = 'baseline')
plt.legend();
plt.xlabel('False Positive Rate');
plt.ylabel('True Positive Rate'); plt.title('ROC Curves');
plt.show();
metrics.confusion_matrix(did_stock_go_up_actual, did_stock_go_up_predicted)
```
