```python
#Question 1
import pandas as pd
df = pd.read_csv("fetal_health.csv")
df
```

```python
train = df.head(n=1500)
train.describe()
```

```python
test = df.iloc[1500:]
test.describe()
```

```python
from sklearn import tree
clf=tree.DecisionTreeClassifier()
train_X = train[train.columns[:-1]]
train_X.head()
```

```python
train_Y = train['fetal_health']
train_Y.head()
```

```python
clf.fit(train_X,train_Y)
```

```python
test_X = test[test.columns[:-1]]
test_predict_Y=clf.predict(test_X)
test_predict_Y
```

```python
from sklearn.metrics import confusion_matrix
test_Y = test['fetal_health']
cf = confusion_matrix(test_Y,test_predict_Y) 
cf # IN FORMAT:
# |            | Predicted 1.0 | Predicted 2.0 | Predicted 3.0 |
# |Actually 1.0|
# |Actually 2.0|
# |Actually 3.0|
```

```python
#Let's turn this 3x3 confusion matrix into a 2x2 using 1.0 as our Positive class.
cf_condensed = [[cf[0][0], (cf[0][1] + cf[0][2])],
                [(cf[1][0]+cf[2][0]),(cf[1][1]+cf[1][2]+cf[2][1]+cf[2][2])]]
FP = cf_condensed[1][0]
FN = cf_condensed[0][1]
TP = cf_condensed[0][0]
TN = cf_condensed[1][1]
print(f" | {TP} | {FN} | " + "\n" + f" | {FP}  | {TN}  |")
```

```python
from sklearn import metrics
sk_precision = metrics.precision_score(test_Y, test_predict_Y, average=None)
sk_precision
```

```python
cf_precision = TP / (TP + FP)
cf_precision 
#Note that this matches the precision of Class 1 in sk_precision,
#because our cf counts Class 1 as the positive class.
```

```python
sk_recall = metrics.recall_score(test_Y, test_predict_Y, average=None)
print(sk_recall)
cf_recall = TP / (TP + FN)
print(cf_recall)
```

```python
sk_MCC = metrics.matthews_corrcoef(test_Y, test_predict_Y)
sk_MCC #Think I'll skip the "my own code" for this one.
```

```python
#QUESTION 2
names_col=['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','income']
train_df = pd.read_csv('adult.data',header=None,names=names_col)
test_df = pd.read_csv('adult.test',header=None,names=names_col)
train_df.describe()
```

```python
train_X = train_df[train_df.columns[:-1]]
train_Y = train_df['income']
test_X = test_df[test_df.columns[:-1]]
test_Y = test_df['income']
train_X.dtypes
```

```python
test_Y_binary = [] #A lot of these metrics really don't like non-binaries, so <=50k is negative, >50k is positive.
for value in test_Y:
    if(value == ' <=50K.'):
        test_Y_binary.append(0)
    else:
        test_Y_binary.append(1)
train_Y_binary = []
for value in train_Y:
    if(value == ' <=50K'):
        train_Y_binary.append(0)
    else:
        train_Y_binary.append(1)
```

```python
features_to_encode = list(train_X.select_dtypes(include=['object']).columns)
features_to_encode
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.ensemble import RandomForestClassifier
col_trans = make_column_transformer(
                        (OneHotEncoder(handle_unknown = "ignore"),features_to_encode),
                        remainder = "passthrough"
                        )
rf_classifier = RandomForestClassifier(n_estimators=100)
from sklearn.pipeline import make_pipeline
clf = make_pipeline(col_trans, rf_classifier)
clf.fit(train_X,train_Y_binary)
```

```python
test_Y_predicted = clf.predict(test_X)
test_Y_predicted
```

```python
train_probs = clf.predict_proba(train_X)[:,1]
probs = clf.predict_proba(test_X)[:,1]

train_predictions = clf.predict(train_X)
print(f'Train ROC AUC Score: {metrics.roc_auc_score(train_Y_binary, train_probs)}')
print(f'Test ROC AUC  Score: {metrics.roc_auc_score(test_Y_binary, probs)}')
```

```python
fpr, tpr, threshholds = metrics.roc_curve(test_Y_binary,test_Y_predicted,pos_label=1)
base_fpr, base_tpr, _ = metrics.roc_curve(test_Y_binary, [1 for _ in range(len(test_Y_binary))])
plt.figure(figsize = (8, 6))
plt.rcParams['font.size'] = 16
plt.plot(fpr, tpr, 'r', label = 'model')
plt.plot(base_fpr, base_tpr, 'b', label = 'baseline')
plt.legend();
plt.xlabel('False Positive Rate');
plt.ylabel('True Positive Rate'); plt.title('ROC Curves');
plt.show();
```

```python
metrics.plot_roc_curve(clf,test_X,test_Y_binary) #I guess this one is more accurate
plt.show()     
```

```python
from sklearn.model_selection import GridSearchCV
hyperparameters = { 'randomforestclassifier__max_features' : ['auto', 'sqrt', 'log2'],
                  'randomforestclassifier__max_depth': [3,5,7,10]}
clf = make_pipeline(col_trans, rf_classifier)
clftuned = GridSearchCV(clf, hyperparameters)
clftuned.fit(train_X, train_Y_binary)
predtuned = clf.predict(test_X)
print(f'F1 Score: {metrics.f1_score(test_Y_binary,predtuned)}') #Best score: 0.65
```

```python
#QUESTION 3
df = pd.read_csv("superconductors.csv")
df
```

```python
train_df = df.head(n=(int(len(df.index)*0.8)))
test_df = df.tail(n=(int(len(df.index)*0.2)))
train_X = train_df[train_df.columns[:-1]]
train_Y = train_df['critical_temp']
test_X = test_df[test_df.columns[:-1]]
test_Y = test_df['critical_temp']
```

```python
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel as C
import numpy as np

#np.random.seed(1)
#kernel = C(1.0, (1e-3, 1e3)) * RBF(10, (1e-2, 1e2))
gp = GaussianProcessRegressor()
gp.fit(train_X,train_Y)
y_pred, sigma = gp.predict(test_X, return_std=True) # This takes forever to run (this thing scales O(n^3)?!)
#thank you for giving us the biggest data set in human history to use with the least efficent algorithm
print(sigma)
y_pred
```

```python
#Plot the parity plot:
plt.scatter(test_Y, y_pred)
plt.plot(test_Y,test_Y,'k-') # identity line
font_axis_publish = {
        'color':  'black',
        'weight': 'bold',
        'size': 16}
plt.title("Parity Plot", fontdict=font_axis_publish)
plt.xlabel('True Temperature', fontdict=font_axis_publish)
plt.ylabel('Prediction', fontdict=font_axis_publish)
plt.show()
```
